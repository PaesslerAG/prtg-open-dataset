# Visualizaion of 50 examples from the dataset

```python
%matplotlib ipympl
import requests
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [10, 6]


root = 'https://raw.githubusercontent.com/PaesslerAG/prtg-open-dataset/main'

def get_all_examples():
    r = requests.get(str.format(F'{root}/metadata/examples.txt'))
    return r.content.decode().split('\n')[:-1]

for file in get_all_examples():
    df = pd.read_csv(str.format(F'{root}/examples/{file}'), delimiter='\t', parse_dates=['ts'])
    plt.figure(file)
    plt.scatter(df.ts.tolist(), df.iloc[:,-1].tolist(),s=1,c='k')
    plt.yscale('log')
    plt.show()
```
